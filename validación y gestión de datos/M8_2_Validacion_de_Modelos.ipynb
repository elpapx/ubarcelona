# 1. Validación de modelos


La gran diferencia entre Probabilidad y Estadistica es que en la probabilidad conocemos el modelo y solemos buscar el cálculo de la variable dependiente (salida). Mientras que, en la estadistica, no conocemos el modelo sino que conocemos las variables independientes y dependientes para hallar el modelo. 

<img src="https://raw.githubusercontent.com/pifparfait/IL3/main/Val_modelo.png?token=AGKQN32XS7ZABG7QBQIHWRLA5HUOC" style="max-width:100%;"/>

En el caso de la estadistica, puede ser que no tengamos todas las variables, pero tenemos una muestra de la población. Por lo que, buscar el modelo, sería ver si nuestros datos son los más representativos para generalizar nuestro modelo o no. Aquí el estudio del error será clave.

<img src="https://raw.githubusercontent.com/pifparfait/IL3/main/ML.png?token=AGKQN34NACSMJHWML7NCC4TA5HURY" style="max-width:100%;"/>



Elegir el modelo de regresión lineal correcto puede resultar complicado. Intentar modelarlo con solo una muestra no lo hace más fácil. 
Durante esta sesión, revisaremos algunos métodos estadísticos comunes para seleccionar modelos, las complicaciones que puede enfrentar y, brindaremos algunos consejos prácticos para elegir el mejor modelo de regresión.

Desde el punto de vista de un matemático: se investiga para buscar la relación entre algunos predictores y la variable de respuesta. 

Desde el punto de vista de un investigador: se investiga midiendo muchas variables, pero solo se incluyen algunas de ellas en el modelo. 

Desde el punto de vista del analista:
Por el hecho del coste computacional, el analista busca la eficiencia intentando eliminar las variables que no están relacionadas e incluye solo aquellas con una verdadera relación. En el camino, los analistas consideran muchos modelos posibles.

Se esfuerzan por lograr un [principio de Ricitos de Oro o Goldilocks principle](https://es.vvikipedla.com/wiki/Goldilocks_principle) con la cantidad de predictores que incluyen.

>NOTAS:

**Muy pocos**: un modelo subespecificado tiende a producir estimaciones sesgadas.

**Demasiados**: un modelo sobreespecificado tiende a tener estimaciones menos precisas.

**Justo a la derecha**: un modelo con los términos correctos no tiene sesgos y las estimaciones más precisas.

### 1.1 En caso de la regresión:

**R-cuadrado ajustado** y **R-cuadrado previsto**: generalmente, uno elige los modelos que tienen valores de R-cuadrado ajustados y pronosticados más altos. Estas estadísticas están diseñadas para evitar un problema clave con **R-cuadrado regular** que aumenta cada vez que agrega un predictor y puede llegar a hacer que el modelo sea demasiado complejo.

**R cuadrada ajustada** aumenta solo si el nuevo término mejora el modelo más de lo que se esperaría por casualidad y también puede disminuir con predictores de mala calidad.

**R-cuadrado predicho** es una forma de validación cruzada (*cross-validation*) y también puede disminuir. La *validación cruzada* determina qué tan bien se generaliza su modelo a otros conjuntos de datos mediante la partición de sus datos.

**Valores p** para los predictores: en la regresión, los *valores p* bajos indican términos que son estadísticamente significativos. “Reducir el modelo” se refiere a la práctica de incluir todos los predictores candidatos en el modelo y luego eliminar sistemáticamente el término con el *valor p* más alto uno por uno hasta que se quede solo con predictores significativos.

#### 1.2.1 La curva ROC, gráfico de ganancias y curva Lift

El nombre de la curva *ROC-AUC* proviene de las siglas de Area Under Curve y Receiver Operating Characteristic. 

<img src="https://raw.githubusercontent.com/pifparfait/IL3/main/ROC.png?token=AGKQN36XDIBDQU652IWZLA3A5HVHG" style="max-width:100%;"/>

Para construir la curva ROC-AUC, se necesitan dos medidas para calcular la Matriz de confusión, dicho en otras palabras, la tasa de verdaderos positivos (o recuperación) y la tasa de falsos positivos (caída). Para ello, trazaremos TPR en el eje y y FPR en el eje x para los distintos umbrales en el rango [0,1]. Recuerda que, reducir o aumentar el umbral tiene un gran impacto en la capacidad de nuestro modelo para predecir las clases negativas y positivas.

<img src="https://github.com/pifparfait/IL3/blob/main/confusion-matrix.png?raw=true" style="max-width:100%;"/>

Fuente: https://en.wikipedia.org/wiki/Template:Diagnostic_testing_diagram

\\

Comencemos explicando una convención que muchas veces genera confusión.

Un *clasificador binario tiene dos salidas: clase 1 y clase 2*.

Ahora, según el dominio aplicado, esto podría ser una salida sí / no, una salida 0/1, una salida buena / mala o cualquier otra cosa que tenga dos valores.

Se ha de tener mucho cuidado con lo que se entiende como Condición Prevista Positiva y como Condición Prevista Negativa. Todo se reduce a lo que se define como clase positiva y, no necesariamente, tiene un significado "positivo" como en el lenguaje hablado. Por ejemplo: en una predicción Sin incumplimiento / Incumplimiento, para una solicitud de préstamo, una Condición Prevista Positiva sería el caso en el que el solicitante incumple con su préstamo, mientras que, una Condición Prevista Negativa vendría a ser cuando el solicitante puede reembolsar su préstamo. Otro ejemplo podría ser cuando se realiza una prueba de cáncer. La Condición Prevista Positiva sería cuando se sabe que el paciente tiene cáncer, mientras que la Condición Prevista Negativa sería el caso en el que el paciente no tiene cáncer. Todo se reduce a lo que se predice.

Entonces, cuando se obtienen los resultados de un clasificador, se quiere que, muchas de las predicciones estén en la diagonal principal. La celda superior izquierda indica cuántas de las predicciones positivas fueron realmente positivas (Verdadero Positivo) y la celda inferior izquierda muestra cuántas de las predicciones negativas fueron realmente negativas (Verdadero Negativo). Hasta aquí todo bien.

La esquina superior derecha son los falsos positivos. Este número muestra cuántas instancias se predijeron como Positivas pero en realidad fueron negativas (llamado **error de Tipo I** en las estadísticas). La celda inferior izquierda de la matriz muestra cuántas instancias se predijeron como negativas pero, en realidad, fueron positivas (**error de tipo II** si alguien pregunta en una entrevista).



<a id=roc_in></a>

#### 1.2.2 Las combinaciones
Procedamos ahora con las combinaciones de estas cuatro celdas y comentemos el valor de cada una por separado.

**Precisión Accuracy**: dice qué tan bien lo hicimos en general. Es la suma de todas las predicciones exitosas divididas entre el número total de observaciones. En términos matemáticos:

$Acc = \frac{(TP + TN)}{(TP + TN + FP + FN)}$

Por lo que, la mayoría de las veces, ésta no es una buena métrica.

**Recall**: nos dice cuántas de las observaciones positivas en los datos pudimos predecir correctamente. Por lo tanto:

$Recall = \frac{TP} {(TP + FN)}$

Recuerda que los falsos negativos son en realidad positivos que clasificamos erróneamente como negativos. La recuperación también se conoce como Tasa Positiva Verdadera o Sensibilidad.

**Precisión**: nos informa de cuántos de los ejemplos que predijimos como Positivos pertenecen realmente a la clase positiva. En términos matemáticos:

$Precisión =\frac{TP}{(TP + FP)}$

Recuerda que los falsos positivos son observaciones que pronosticamos como positivas, pero que, en realidad, son negativas.

**Especificidad**: nos informa de cuántas instancias negativas pudimos predecir correctamente entre todos los ejemplos negativos reales en los datos. Por tanto:

$ Spec =\frac {TN}{(TN + FP)}$

Recuerda que los falsos positivos son en realidad ejemplos negativos, clasificados incorrectamente como positivos. Puedes encontrar este término también como Tasa Negativa Verdadera.

**Fall-out**: se mide como:

$\frac {FP}{(FP + TN)}$

O, si ya se ha calculado la Especificidad, se mide como:

$1-TNR$

La caída también se conoce como *tasa de falsos positivos*.

**F1-Score**: hay casos en los que diferentes modelos tendrán rendimientos similares pero con diferentes puntuaciones de recuperación y precisión. Un modelo puede hacerlo mejor en Recall, y otro en Precisión. Entonces, ¿cuál es mejor? esta medida es en realidad la media armónica entre Precision y Recall que nos proporciona una herramienta para comparar diferentes modelos usando una única métrica. Puede medirse en una variedad de formas algéricas, siendo mi favorita:

$F1-Score = \frac{2 * TP}{(2 * TP + FP + TN)}$

Hay otras métricas para el analisis de modelos de **clasificación del árbol de decisión** que analizaremos a través de una tarea guiada.

## 1.3 La realidad o complicaciones de los modelos en la vida cotidiana
Hay una variedad de métodos estadísticos para ayudarnos a elegir el mejor modelo. Desafortunadamente, también hay una serie de posibles complicaciones. ¡Tenemos que buscar el equilibrio en la fuerza!

> - El mejor modelo puede ser tan bueno como las variables medidas por el estudio. ***Los resultados de las variables que se incluyen pueden estar sesgados por las variables significativas que no se incluyen.***

> - La muestra puede ser inusual, ya sea por casualidad o por metodología de recopilación de datos. Los falsos positivos y los falsos negativos son parte del juego cuando se trabaja con muestras.

> - Los valores $p$ pueden cambiar según los términos específicos del modelo. En particular, la multicolinealidad puede minar la importancia y dificultar la determinación del papel de cada predictor.

> - Si se evalúan suficientes modelos, encontraremos variables que parecen ser significativas pero que solo están correlacionadas por casualidad. Esta forma de minería de datos puede hacer que los datos aleatorios parezcan significativos. Un R-cuadrado de predicción baja es una buena forma de comprobar este problema.

> - Los valores $P$, el R-cuadrado predicho y ajustado y, el $Cp$ de Mallows pueden sugerir diferentes modelos.
La regresión por pasos y la regresión de los mejores subconjuntos son excelentes herramientas y pueden acercarnos al modelo correcto. Sin embargo, los estudios han encontrado que, generalmente, no eligen el modelo correcto.

# 2. Recomendaciones para encontrar el mejor Modelo de Regresión 
Elegir el modelo de regresión correcto puede llegar a ser un arte. Es bueno saber que los métodos estadísticos son de gran ayuda para detectar la dirección correcta aunque, en última instancia, deberán incorporarse otras consideraciones.

**Teoría**

Investiga lo que otros han hecho e incorpora esos hallazgos al construir tu modelo.

Antes de comenzar el análisis de regresión, desarrolla una idea de cuáles son las variables importantes junto con sus relaciones, signos de coeficientes y magnitudes de efecto. Básate en los resultados de otros y facilita tanto la recopilación de los datos correctos como la especificación del mejor modelo de regresión sin la necesidad de minería de datos.

Las consideraciones teóricas no deben descartarse basándose únicamente en medidas estadísticas. Después de ajustar tu modelo, determina si se alinea con la teoría y realiza los ajustes que sean necesarios. Por ejemplo, según la teoría, puedes incluir un predictor en el modelo incluso si su *valor p* no es significativo. Si alguno de los signos de coeficiente contradice la teoría, investiga y cambia tu modelo o explica la inconsistencia.

**Complejidad**

Puedes pensar que los problemas complejos requieren modelos complejos pero, muchos estudios muestran que los modelos más simples, generalmente, producen predicciones más precisas. Dados varios modelos con capacidad explicativa similar, es más probable que el más simple sea la mejor opción. Comienza de manera simple y solo haz que el modelo sea más complejo según sea necesario. Cuanto más complejo sea tu modelo, más probable es que estés adaptando el modelo a tu conjunto de datos de forma específica, por lo que, la generalización sufre.

Verifica que la complejidad agregada realmente produzca intervalos de predicción más estrechos. *¡Verifica el R-cuadrado predicho y no persigas sin pensar en un R-cuadrado alto regular!*

**Parcelas residuales**

A medida que evalúas los modelos, verifica las *gráficas de residuos* porque pueden ayudarte a evitar modelos inadecuados y a ajustar tu modelo para obtener mejores resultados. Por ejemplo, el sesgo en modelos subespecificados puede aparecer como patrones en los residuos, como la necesidad de modelar la curvatura. El modelo más simple que produce residuos aleatorios es un buen candidato para ser un modelo relativamente preciso e insesgado.

> *Recuerda que ninguna medida puede decirte qué modelo es el mejor*. Los métodos estadísticos no comprenden el proceso subyacente o el área temática. *¡El conocimiento es una parte crucial del proceso!*

La regresión lineal calcula una ecuación que minimiza la distancia entre la línea ajustada y todos los puntos de datos. Técnicamente, la regresión de mínimos cuadrados ordinarios (MCO) minimiza la suma de los residuos cuadrados.

>**¿Ves Patrones NO aleatorios en los residuos de tu modelo?**

Si ves patrones no aleatorios en tus residuos, significa que a tus predictores les falta algo.

Siempre hay que dar una justificación completa en el patrón de los residuos de tu modelo, y dejar claro por qué se necesita aleatoriedad en tus residuos. Debes explicar todo lo que sea posible con tus predictores para que solo quede un error aleatorio.

> ¿Por qué? para empezar, analicemos y definamos los 2 componentes básicos de un modelo de regresión válido:
> -  Salida / Respuesta = (Constante + Predictores) + Error

puede ser visto como:

> - Salida / Respuesta = determinista + estocástico

**Parte determinista**

El valor esperado de la respuesta es una función de un conjunto de variables predictoras. Toda la información explicativa / predictiva del modelo debe estar en esta parte.

**Parte estocástico**

Estocástico, según la RAE significa *aleatorio e impredecible*. **El error es la diferencia entre el valor esperado y el valor observado**. Juntando esto, las diferencias entre los valores esperados y observados deben ser impredecibles. En otras palabras, ninguna de la información explicativa / predictiva debe estar en el error.

La idea es que la parte determinista de tu modelo sea tan buena de explicar que, para predecir su respuesta, solo quede la aleatoriedad inherente de cualquier fenómeno del mundo real para la parte del error. Si observas poder explicativo o predictivo en el error, sabrás que a tus predictores les falta parte de la información predictiva. *¡Las parcelas residuales te ayudarán a comprobar esto!*

### 2.3.3 R-cuadrado o Coeficiente de determinación

El coeficiente de determinación mide cuánta de la variación de $y$ es explicada por el modelo. Se calcula como:

$$R^2 = 1 - \frac{\sigma_e^2}{\sigma_y^2}$$

Si la varianza de los errores o residuales $\sigma_e^2$ es cero, el modelo explica el 100% de la variable $y$. Si $\sigma_e^2$ es igual a la varianza de $y$, el modelo no explica nada y $R^2$ vale cero.

$R^2$ determina la calidad del modelo para replicar los resultados, y la proporción de variación de los resultados que puede explicarse por el modelo.

### 2.3.4. R-cuadrado ajustado o Coeficiente de correlación ajustado

El coeficiente de correlación ajustado $\bar{R}^2$ corrige el valor de $R^2$ por la cantidad de variables $k$ (igual a 2 para el caso analizado) y la cantidad de datos $N$:

El modelo es más creíble si hay más datos, y menos creible en caso contrario.

### 2.4.1 Analisis de los residuos Jarque-Bera, Chi^2 two-tail prob, Skew, Kurtosis

En cuanto a cada prueba de hipótesis, es necesario plantear una hipótesis nula para validar:

* $ H_0$: No hay cambios en los datos.
* $ H_1$: Si que hay cambios en los datos.



<a id=JB></a>
**2.4.1.1 Jarque-Bera**

La variable Jarque-Bera está escrita cómo:
$$
    \mathit{JB} = \frac{n-k}{6} \left( S^2 + \frac14 (K-3)^2 \right)
$$

con:
* $n$, es el número de observaciones.
* $k$, es el número de variables explicativas si los datos provienen de los residuos de una regresión lineal. De lo contrario, $k$ permanece en cero.
* $S$, es el  coeficiente de asimetría de la muestra analizada (estadística).
* $K$, es la curtosis de la muestra analizada.

Matemáticamente,  $S$  y $K$ se definen por:

$$
    S = \frac{\hat {\mu}_3}{\hat {\sigma}^3}
        = \frac {\frac {1}{n} \sum_{i = 1}^ n (x_i- \bar{x})^ 3} {(\frac {1}{n} \sum_{i = 1}^ n(x_i- \bar {x}) )^ 2 \ )^{3/2}} 
$$

\\
$$
     K = \frac {\hat {\mu}_4} {\hat {\sigma}^ 4}
        = \frac {\frac{1}{n} \sum_{i = 1}^ n (x_i- \bar{x})^ 4} { (\frac{1}{n} \sum_{i = 1}^ n(x_i- \bar{x}) )^2  )^2}
$$

con $\hat{\mu}_3 $ y $\hat{\mu}_4$ como los estimadores del tercer y cuarto momento, $\bar{x}$ es la media de la muestra y $ \hat {\sigma}^ 2$ es la varianza de la muestra (estadísticas y probabilidades).

La estadística "JB" sigue asintóticamente una ley de $\chi^{2}$ con dos grados de libertad.

Esta prueba se utiliza, con frecuencia, para determinar si los residuos de una regresión lineal múltiple y de una regresión lineal, siguen una distribución normal. 




<a id=chi></a>
**2.4.1.2 Chi cuadrado**

En este caso, $N$ observaciones se dividen en $n$ celdas. Un ejemplo muy claro de este hecho es que en una población, un determinado suceso sucede en cada celda en una determinada frecuencia. La "frecuencia teórica" de cada celda se calcula como:

$$
E_{i}=N/n 
$$

Los grados de libertad de la prueba de la bondad de ajuste $\upsilon$ se calculan como: $\upsilon = n-s$, donde $n$ es el número de celdas, o donde $s$ es el número de parámetros estadísticos que hemos tenido que suponer como poblacionales siendo muestrales.

El valor del estadístico de la prueba es

$$
\chi^{2} = \sum_{i=1}^{n} {(O_i - E_i)^2 \over E_i}
$$

donde:

* $\chi^{2}$ es el estadístico de la prueba que se acerca asintóticamente a una distribución $\chi^{2}$.

* $O_{y}$ es la frecuencia observada.

* $E_{y}$ es la frecuencia teórica calculada según la distribución de la hipótesis nula.

* $n$ es el número de celdas.




<a id=curt></a>
**2.4.1.3 Curtosis o Kurtosis**

La curtosis es el cuarto momento estandarizado, definido como:

$$
\operatorname{Kurt}[X] = \operatorname{E}\left[\left(\frac{X - \mu}{\sigma}\right)^4\right] = \frac{\operatorname{E}\left[(X - \mu)^4\right]}{\left(\operatorname{E}\left[(X - \mu)^2\right]\right)^2} = \frac{\mu_4}{\sigma^4}
$$

donde:

*  $ \mu^4$ es el cuarto momento central y 
*  $\sigma$ es la desviación estándar.

En la literatura, se utilizan varias letras para denotar la curtosis. Una opción muy común es $\chi$, que está bien siempre que esté claro que no se refiere a un acumulativo.

```python
# Normalidad de los residuos
name = ["Jarque-Bera", "Chi^2 two-tail prob.", "Skew", "Kurtosis"]
test = sms.jarque_bera(results.resid)
lzip(name, test)
```

# 3.  Implementación de un Clasificador Binario
Para implementar un clasificador binario, deben seguirse los siguientes pasos:

En primer lugar, importar los modulos complementarios:


```python
sns.set_style("whitegrid")
%matplotlib inline
```

En segundo lugar, importar la base de datos. En este caso, trabajaremos con el **dataset bank-full**:

```python
df = pd.read_csv('./dataset/bank-full.csv')
```

En tercer lugar, encoder las variables categóricas:

```python
print('Conversión de variables categóricas en numéricas...')

categorical_columns = ['job', 'marital', 'education', 'default', 'housing',
'loan', 'contact', 'day', 'month', 'campaign',
'previous', 'poutcome']

use_cols = ['month', 'campaign', 'Target']
df['Target'] = df['Target'].map({'yes':1, 'no':0})
 
for c in categorical_columns: 
    df[c] = df[c].astype("category")
    df[c] = df[c].cat.codes
```

En cuarto lugar, montar los bancos de pruebas y de entrenamiento de la siguiente forma:
- Dividiendo en un 75% de entrenamiento y un 25% de prueba.
- Creando la etiqueta de destino separada.
- Creando la etiqueta de destino de caída.

```python
#Dividir en 75% de tren y 25% de prueba
train_df, test_df = train_test_split(df, test_size = 0.25, random_state= 1984)
 
# Etiqueta de destino separada
train_y = train_df['Target']
test_y = test_df['Target']
 
# Etiqueta de destino de caída
train_x = train_df.drop(['Target'], axis = 1)
test_x = test_df.drop(['Target'], axis = 1)
```

En quinto lugar, construir el modelo Random Forest y ajustar los modelos:

```python
#Build Random Forest model
model1 = RandomForestClassifier(max_depth=4, n_estimators = 20)
model1.fit(train_x,train_y)
model2 = RandomForestClassifier(max_depth=12, n_estimators = 120)
model2.fit(train_x,train_y)
```

Y, en sexto y último lugar, predecir el conjunto de pruebas:

```python
#Predict test set
pred1 = model1.predict_proba(test_x)
pred2 = model2.predict_proba(test_x)
```

## 4.1 Algoritmo del árbol de decisión

Un árbol de decisión es una estructura de árbol similar a un diagrama de flujo donde un nodo interno representa una característica también conocida como atributo.
Normalmente, la rama representa una regla de decisión y cada nodo hoja representa el resultado. 

El nodo superior de un árbol de decisiones se conoce como nodo raíz o *padre*. El nodo padre aprende a particionar sobre la base del valor del atributo. Poco a poco divide el árbol de una manera recursiva llamada **partición recursiva**. 

> Esta estructura similar a un diagrama de flujo nos ayuda en la toma de decisiones. *Puede visualizarse como un diagrama de flujo que imita, fácilmente, el pensamiento a nivel humano intuitivo y de fácil interpretación*.

## 4.2 ¿Cómo funcionan?

La idea básica detrás de cualquier algoritmo de árbol de decisión es la siguiente:

 1. Selecciona el mejor atributo utilizando Medidas de selección de atributos (ASM) para dividir los registros.

 2. Convierte ese atributo en un nodo de decisión y divide el conjunto de datos en subconjuntos más pequeños.

 3. Inicia la construcción del árbol repitiendo este proceso de forma recursiva para cada "hijo" hasta que una de las condiciones coincida:

  - Todas las tuplas pertenecen al mismo valor de atributo.

  - No quedan más atributos.

  - No hay más instancias.

## 4.3 Medidas de selección de atributos

La medida de selección de atributos es una heurística para seleccionar el criterio de división que divide los datos de la mejor manera posible.

También se conoce como reglas de división porque nos ayuda a determinar los puntos de interrupción de las tuplas en un nodo determinado. ASM proporciona un rango a cada característica / atributo al explicar el conjunto de datos dado. 

El atributo de mejor puntuación se seleccionará como un atributo de división. 

En el caso de un atributo de valor continuo, también es necesario definir los puntos de división para las ramas. Las medidas de selección más populares son la ganancia de información, la relación de ganancia y el índice de Gini.

### 4.3.1 Ganancia de información  

El concepto de entropía, una herencia de Shannon, mide la impureza del conjunto de entrada. En física y matemáticas, la entropía se refiere a la aleatoriedad o la impureza en el sistema. En teoría de la información, se refiere a la impureza en un grupo de ejemplos. **La ganancia de información es la disminución de la entropía**. La ganancia de información calcula la diferencia entre la entropía antes de la división y la entropía promedio después de la división del conjunto de datos en función de los valores de atributo dados. El algoritmo del árbol de decisión ID3 (dicotomizador iterativo)utiliza la ganancia de información:

$$\textrm{Info}(D) = -\sum_{i = 1}^{m} p_i\log_2 p_i$$



Donde, $P_{i}$ es la probabilidad de que una tupla arbitraria en $D$ pertenezca a la clase $C_{i}$.

$$\textrm{Info}_A(D) = \sum_{j = 1}^{V} \dfrac{|D_j|}{|D|} \textrm{ x }\textrm{Info}(D_j)$$

$$\textrm{Gain}(A) = \textrm{Info}(D) - \textrm{Info}_A(D)$$


Donde:

- **Info(D)** es la cantidad promedio de información necesaria para identificar la etiqueta de clase de una tupla en D.

- **|$D_{j}$|/| $D$|** actúa como el peso de la j-ésima partición.

- **InfoA(D)** es la información esperada requerida para clasificar una tupla de D según la partición por A.

El atributo A con la mayor ganancia de información, **Ganancia (A)**, se elige como atributo de división en el nodo $N()$.

### 4.3.2 Ratio de ganancia

La obtención de información está sesgada para el atributo con muchos resultados. Esto significa que prefiere el atributo con una gran cantidad de valores distintos. 

> Un caso podría ser el atributo con un identificador único como *customer_ID* que tiene cero información (D) debido a la partición pura. Esto maximizaría la ganancia de información y crearía particiones inútiles.

C4.5, una mejora de ID3, utiliza una extensión de la ganancia de información conocida como relación de ganancia. La relación de ganancia maneja el problema del sesgo al normalizar la ganancia de información usando *Split Info*. La implementación de Java del algoritmo C4.5 se conoce como J48, que está disponible en la herramienta de minería de datos WEKA:

$$ \textrm{SplitInfo}_A(D) = - \sum_{j = 1}^{V} \dfrac{|D_j|}{|D|} \textrm{ x } log_2(\frac{|D_j|}{|D|})$$

Donde:
- $|D_{j} | / | D |$ actúa como el peso de la $j$-ésima partición.

- $v$ es el número de valores discretos en el atributo $A$.

La relación de ganancia se puede definir como:

$$ \textrm{GainRatio}(A) = \dfrac{\textrm{Gain(A)}}{\textrm{SplitInfo}_A(D)} $$
El atributo con la relación de ganancia más alta se elige como atributo de división.

### 4.3.3 Índice de Gini

Otro algoritmo de árbol de decisión CART (Árbol de clasificación y regresión) utiliza el método Gini para crear puntos de división:

$$Gini(D) = 1 - \sum_{i = 1}^{m} P_i^2 $$

Donde, $P_{i}$ es la probabilidad de que una tupla en $D$ pertenezca a la clase $C_{i}$.

<img src="https://github.com/pifparfait/IL3/blob/main/tree.png?raw=true" style="max-width:100%;"/>

El índice de Gini considera una división binaria para cada atributo. Puede calcular una suma ponderada de la impureza de cada partición. Si una división binaria en el atributo $A$ divide los datos $D$ en $D_{1}$ y $D_{2}$, el índice de Gini de $D$ es:

$$Gini_A(D) = \frac{|D_{1}|}{D} Gini(D_{1}) + \frac{D_{2}}{D} Gini(D_{2})$$

En el caso de un atributo de valor discreto, el subconjunto que da el índice de Gini mínimo para el elegido se selecciona como un atributo de división. En el caso de atributos con valores continuos, la estrategia consiste en seleccionar cada par de valores adyacentes como un posible punto de división y, elegir el punto con un índice de Gini más pequeño como punto de división:

$$ \Delta Gini(A) = Gini(D) - Gini_A(D)$$

## 4.6 Analisis de poda
Para aplicar el proceso de *pruning*, es necesario indicar el argumento *ccp_alpha* que determina el grado de penalización por complejidad. Cuanto mayor es este valor, más agresivo es el podaje y menor el tamaño del árbol resultante. Dado que no hay forma de conocer de antemano el valor óptimo de *ccp_alpha* se recurre a validación cruzada para identificarlo.

```python
# Pruning (const complexity pruning) por validación cruzada
# ------------------------------------------------------------------------------
# Valores de ccp_alpha evaluados
param_grid = {'ccp_alpha':np.linspace(0, 80, 20)}

# Búsqueda por validación cruzada
grid = GridSearchCV(
        # El árbol se crece al máximo posible para luego aplicar el pruning
        estimator = DecisionTreeRegressor(
                            max_depth         = None,
                            min_samples_split = 2,
                            min_samples_leaf  = 1,
                            random_state      = 123
                       ),
        param_grid = param_grid,
        cv         = 10,
        refit      = True,
        return_train_score = True
      )

grid.fit(X_train, y_train)

fig, ax = plt.subplots(figsize=(6, 3.84))
scores = pd.DataFrame(grid.cv_results_)
scores.plot(x='param_ccp_alpha', y='mean_train_score', yerr='std_train_score', ax=ax)
scores.plot(x='param_ccp_alpha', y='mean_test_score', yerr='std_test_score', ax=ax)
ax.set_title("Error de validacion cruzada vs hiperparámetro ccp_alpha");

```

```python
# Mejor valor ccp_alpha encontrado
grid.best_params_
```

> Una vez identificado el valor óptimo de <code>ccp_alpha</code>, se reentrena el árbol indicando este valor en sus argumentos. Si en el <code>GridSearchCV()</code>, se indica <code>refit=True</code>, el reentrenamiento se hace automáticamente y el modelo resultante se encuentra almacenado en <code>.best_estimator_</code>.

```python
# Estructura del árbol final
modelo_final = grid.best_estimator_
print(f"Profundidad del árbol: {modelo_final.get_depth()}")
print(f"Número de nodos terminales: {modelo_final.get_n_leaves()}")

fig, ax = plt.subplots(figsize=(7, 5))
plot = plot_tree(
            decision_tree = modelo_final,
            feature_names = feature_cols,
            class_names   = ["0", "1"],
            filled        = True,
            impurity      = False,
            ax            = ax
       )
```

## Vamos a entrenar nuestro árbol de decisión para el problema de regresión
La clase DecisionTreeRegressor del módulo sklearn.tree permite entrenar árboles de decisión para problemas de regresión. A continuación, se ajusta un árbol de regresión empleando como variable respuesta PRICE y como predictores todas las otras variables disponibles.

DecisionTreeRegressor tiene por defecto los siguientes hiperparámetros:

    criterion='mse'
    splitter='best'
    max_depth=None
    min_samples_split=2
    min_samples_leaf=1
    min_weight_fraction_leaf=0.0
    max_features=None
    random_state=None
    max_leaf_nodes=None
    min_impurity_decrease=0.0
    min_impurity_split=None
    ccp_alpha=0.0
De entre todos ellos, los más importantes son aquellos que detienen el crecimiento del árbol (condiciones de stop):

- **max_depth:** profundidad máxima que puede alcanzar el árbol.

- **min_samples_split:** número mínimo de observaciones que debe de tener un nodo para que pueda dividirse. Si es un valor decimal, se interpreta como fracción del total de observaciones de entrenamiento ceil (min_samples_split * n_samples).

- **min_samples_leaf:** número mínimo de observaciones que debe de tener cada uno de los nodos hijos para que se produzca la división. Si es un valor decimal, se interpreta como fracción del total de observaciones de entrenamiento ceil (min_samples_split * n_samples).

- **max_leaf_nodes:** número máximo de nodos terminales.

- **random_state:** semilla para que los resultados sean reproducibles. Tiene que ser un valor entero.

* Desde el punto de **vista del analista**: por el hecho del coste computacional, el analista busca la eficiencia intentando eliminar las variables que no están relacionadas e incluye solo aquellas con una verdadera relación. En el camino, los analistas consideran muchos modelos posibles.

* Pasos para **validar un modelo**:

  1. Saber si estás en regresión o clasificación.

  2. **Mediar** y **analizar los residuos** ( gráficas y métricas).

  3. **Repetir** varias veces las medidas.

* **Entender la naturaleza del problema**: regresión vs. clasificación. 

* Variables dependientes e independientes, necesidad de **feature engineering**.

* **Entender la precisión** requerida por el problema. 

* Búscar las **soluciones simples**. 

* Usar los **algoritmos con alta capacidad de generalización** (Redes Neuronales, XGBoost u otros Boosted trees).
