# 3 - Estadística Descriptiva Bidimensional

En la parte de estadística descriptiva unidimensional, se analiza todo sobre cada una característica de la población, es decir, una sola variable. En este apartado, se analizarán sobre cada uno de los individuos dos caracteres o variables que se notarán por $X$ e $Y$.

Por ejemplo: estatura y peso de los trabajadores/as de una empresa, edad y peso de los niños, etc.

## 3.1. Distribución de frecuencias bidimensionales

Hasta ahora, se ha estudiado, sobre cada observación de las que forman la muestra, el valor que presenta un determinado carácter. En este punto, se analizan para cada observación dos variables.

En ocasiones recibe el nombre de *análisis bivariante*.

Se debe tener en cuenta que la distribución de frecuencias bidimensional proporciona **tres tipos de información**:

1. Información acerca del patrón de **comportamiento conjunto de ambas variables**.

2. Información acerca del **comportamiento de una de las variables cuando se controla el valor de la otra** (distribuciones condicionadas).  

3. Información **por separado de cada una de las dos variables observadas** (distribuciones marginales).


<font color='Blue'><b> Variable estadística bidimensional </b></font>

Cuando se consideran situaciones en la que el estadístico realiza la observación simultánea de dos caracteres en el individuo, se obtienen pares de resultados.

Los distintos valores de las modalidades que pueden adoptar estos caracteres forman un conjunto de pares, que representamos por $(X, Y)$, y que llamaremos variable estadística bidimensional. Sus valores estarán representados por los pares de valores $(x_{i} , y_{j})$.

**Los dos caracteres observados no tienen por qué ser de la misma clase**, pudiendo presentarse distintas situaciones:
*	Dos caracteres cualitativos: El sexo y el color del pelo de una persona.

* Dos caracteres cuantitativos: El peso y la estatura de una persona.

* Uno cualitativo y otro cuantitativo: La profesión y los años de servicio.

Las variables $(X, Y)$ que representan los valores de dos caracteres cuantitativos y pueden clasificarse de la siguiente manera:

-	X discreta e Y discreta: Número de hijos/as y número de hermanos/as de una persona.

-	X continua e Y continua: Perímetro craneal y perímetro torácico de una persona.

-	X discreta e Y continua: Número de hijos/as y el ingreso económico de una familia.

-	X continua e Y discreta: Temperatura y pulsaciones de una persona.

<br>
<p> <mark>OBSERVACIONES</mark> </p>
<hr>

Los razonamientos que presentaremos para dos variables (estadística bidimensional) son extrapolables en mayor o menor medida para $n$ variables (estadística n-dimensional).

<font color='Blue'><b> Distribución conjunta </b></font>

Se considera una población con $n$ individuos para la que se desea estudiar dos características simultáneamente, $X$ e $Y$, cada una con modalidades $x_{1},x_{2}, x_{3}, \ldots x_{k}$ e $y_{1},y_{2}, y_{3}, \ldots y_{p}$ respectivamente. Entonces, las observaciones correspondientes a cada individuo son el par:

$$ \bbox[5px,border: 2px solid blue]{ 
   (x_{i}, y_{j}) \quad con \quad i = 1,2,3 \ldots  k, \quad j = 1,2,3 \ldots  p \quad siendo \quad k,p \in \Bbb N
} $$

<br>

Se llama **frecuencia absoluta** del par $ (x_{i}, y_{j})$ a $ n_{ij}$ que es  el  número de individuos que presentan, a la vez, la modalidad $x_{i}$  e $y_{j}$. Se representa matemáticamente como $n_{ij}$.

Su fórmula es:

$$ \bbox[5px,border: 2px solid blue]{ 
   (x_{i}, y_{j})= \sum_{i=1}^{k}\sum_{j=1}^{p} n_{ij} = n
} $$

<br>

Se llama **frecuencia relativa** del par $ (x_{i}, y_{j})$ a la proporción de individuos en la población que presentan, simultáneamente, la modalidad $x_{i}$ y la modalidad $y_{j}$. Se representa matemáticamente como $f_{ij}$.

Su fórmula es:

$$ \bbox[5px,border: 2px solid blue]{ 
   f_{ij}= \frac{n_{ij}}{n}
} $$

Donde cumple la propiedad:

$$ \sum_{i=1}^{k}\sum_{j=1}^{p} f_{ij} = \frac{1}{n}\sum_{i=1}^{k}\sum_{j=1}^{p} n_{ij} = 1
$$

<br>

La  distribución  representada por $(x_{i}, y_{j})$ y sus frecuencias absolutas se denomina **distribución conjunta** de $X$ e $Y$. Por tanto, se define la distribución de frecuencias como el conjunto de valores que toma la variable bidimensional, con sus respectivas frecuencias absolutas o relativas.

## 3.2. Tablas estadísticas de doble entrada

**Para ordenar los datos, se utiliza una tabla de doble entrada** donde tengan cabida los $k$ valores distintos de la variable $X$ y los $p$ valores distintos de la variable $Y$. En la tabla se puede expresar el número de veces que se repite para cada par de valores posibles $(x_{i}, y_{j})$, formando el producto cartesiano de los dos conjuntos numéricos.

La t*abla estadística de doble entrada o bidimensional* se representa de la siguiente forma:

| X \  Y | $y_{1}$ | $y_{2}$ | · · · | $y_{j}$ | · · · | $y_{p}$ | $n_{i.}$ |   
|------------|---------|---------|-------|-----------|-------|-----------|-----|
| $x_{1}$ | $n_{11}$| $n_{12}$ | · · · | $n_{1j}$| · · · |$n_{1p}$| $n_{1}$. |
| $x_{2}$ | $n_{21}$| $n_{22}$ | · · · | $n_{2j}$| · · · |$n_{2p}$| $n_{2}$. |
| .  |    .    |    .    |   · · ·     |   .     | · · · |   .    |   .    |
| $x_{i}$ | $n_{i1}$| $n_{i2}$ | · · · | <mark> $n_{ij}$ | · · · |$n_{ip}$| $n_{i}$. |
| .  |    .    |    .    |   · · ·     |   .     | · · · |   .    |
| $x_{k}$ | $n_{k1}$| $n_{k2}$ | · · · | $n_{kj}$| · · · |$n_{kp}$| $n_{k}$. |
| $n_{.j}$ | $n_{.1}$| $n_{.2}$ | · · · | $n_{.j}$| · · · |$n_{.p}$| $n_{ij}$ |

Donde:

* $ n_{i.}$ es la frecuencia absoluta. Representa el número de veces que se ha observado el valor $x_{i}$ de $X$ independientemente del valor presentado por $Y$. Por lo tanto, es la **suma de todas las frecuencias absolutas de la fila** correspondiente al valor $x_{i}$ de $X$.

* $ n_{.j}$ es la frecuencia absoluta. Representa el número de veces que se ha observado el valor $y_{j}$ de $Y$ independientemente del valor presentado por $X$. Por lo tanto, es la **suma de todas las frecuencias absolutas de la columna** correspondiente al valor $y_{j}$ de $Y$.

* $ n_{ij}$ (también conocido como $n$) es la frecuencia absoluta, esto es, es el número de veces que aparece repetido el par $(x_{i}, y_{j})$. Representa el **número total de observaciones**. Por tanto, es el total de las frecuencias absolutas $n_{ij}$ correspondiente a los valores $x_{i}$ de $X$ e $y_{j}$ de $Y$.

<br>
<p> <mark>IMPORTANTE</mark> </p>
<hr>


* Cuando alguna de las variables están agrupadas en intervalos, se toma como valor de $x_{i}$ o $y_{j}$, la marca de clase. 

* En algunos libros recibe el nombre de *tablas de correlación y contingencia*.

<br>

$\bbox{Ejemplo.}$ 

La distribución según salarios (en euros) y edades (en años) de un grupo de 100 personas de entre 20 y 23 años según la tabla que mostramos debajo. 

Analiza los valores de $n_{42}$, el $n_{4.}$ y el $n_{.2}$ de los datos.

|  X \  Y   | [50-100)  ($C_{i} = 75$)  |[100-150)  ($C_{i} = 125$)|     [150-200)  ($C_{i} = 175$) |     $\sum_{fila}$   |
|--------|-----------------|--------------|----------------|---------|
|    20  |      10         |     3        |       2        |     15  |
|    21  |       5         |    15        |       5        |     25  |
|    22  |       2         |    20        |      15        |     37  |
|    23  |       0         |    13        |      10        |     23  |
| $\sum_{columna}$  |   17 |    51        |      32        |     100 |

\\

La interpretación de cada uno de los valores es la siguiente: 

* La frecuencia $n_{42}= 13$ significa que 13 de los 100 jóvenes tienen 23 años y un salario entre 100 y 150 euros. 

* La frecuencia $n_{4.}=n_{41}+n_{42}+n_{43}=0+13+10= 23$ significa que 23 de los 100 individuos tienen 23 años independientemente del salario que tengan.

* La frecuencia $n_{.2}=n_{12}+n_{22}+n_{23}+n_{24}=3+15+20+13= 51$ significa que 51 de los 100 individuos tienen un salario entre 100 y 150 euros independientemente de la edad que tengan.

<br>

La **tabla estadística más sencilla para representar una variable bidimensional** consiste en colocar en dos columnas los pares de valores según hemos ido observando. Un mismo subíndice afecta a ambos elementos del par e indica que observación proporciona dicho par de valores ($x_{i}$, $y_{j}$), el último subíndice será $N$ (el número total de observaciones). 

En esta representación, los distintos valores de la variable $X$  tienen la notación de $x_{i} \; \text{con} \; i= 1, 2,3, \ldots k \;$ y los distintos valores de la variable $Y$ tienen la notación de $y_{j} \; \text{con} \; j= 1, 2, 3, \ldots p$.

A cada observación le corresponden un par de valores $(x_{i}, y_{j})$ y el número de observaciones que han presentado el valor $x_{i}$ de $X$ e $y_{j}$ de $Y$ será la frecuencia absoluta $n_{ij}$ del par $(x_{i}, y_{j})$.

<br>

$\bbox{Ejemplo.}$

La variable $X$ es el peso y la variable $Y$ es la altura.

### 3.2.1. Distribuciones marginales

De estas tablas de doble entrada (de correlación o contingencia),  es  posible extraer la información correspondiente a cada una de las variables (independientemente de la otra). Esta posibilidad es relevante ya que su análisis como variable unidimensional puede ser de utilidad.

A las **distribuciones unidimensionales extraídas de una variable bidimensional** se las denomina distribuciones marginales. Este nombre deriva del hecho de que las frecuencias de la distribución marginal se obtienen sumando en el margen de la derecha o inferior de la tabla de correlación las correspondientes frecuencias bidimensionales.

Dada una tabla de correlación de una variable bidimensional $(X, Y)$, las distribuciones marginales para $X$ e $Y$ serán:

1. Distribución Marginal de $X$:

|     $X$     |  $n_{i}$.   |  $f_{i}$.  |
|-------------|-------------|------------|
| $x_{1}$     |  $n_{1}$.   |  $f_{1}$.  |
| $x_{2}$     |  $n_{2}$.   |  $f_{2}$.  |
|     ....    |             |            |
| $x_{k}$     |   $n_{k}$.  |  $f_{k}$.  |
| $\sum$      |     $N$     |     1      |

2. Distribución Marginal de $Y$:

|     $Y$     |  $n_{i}$.   |  $f_{i}$.  |
|-------------|-------------|------------|
| $y_{1}$     |  $n_{.1}$   |  $f_{.1}$  |
| $y_{2}$     |  $n_{.2}$   |  $f_{.2}$  |
|     ....    |             |            |
| $y_{p}$     |  $n_{.p}$   |  $f_{.p}$  |
| $\sum$      |     $N$     |     1      |

Donde para cada distribución marginal se tiene que:

* La frecuencia absoluta marginal es:

  * $n_{i.}$ es el número de individuos que presentan la modalidad $x_{i}$ del carácter $X$. 

  * $ n_{.j}$ es el número de individuos que presentan la modalidad $y_{j}$ 
del carácter $Y$. 

* La frecuencia relativa marginal es:

  * $ f_{i.} = \frac{n_{i.}}{N} \quad $ que sería equivalente a decir que _i_ es fijo, esto es: $\quad \sum f_{ij} =\sum_{fila} \quad $ 

 * $ f_{.j} = \frac{n_{.j}}{N} \quad $ que sería equivalente a decir que _j_ es fijo, esto es: $\quad \sum f_{ij}=\sum_{columna} \quad$ 

<br>

$\bbox{Ejemplo.}$ 

Continuando con el ejemplo visto en el primer caso, la distribución marginal primera sería:

|     $X$          |    $n_{i.}$  | $f_{i.}$  |
|------------------|--------------|-----------|
|     20           |     15       | 15/100    |
|     21           |     25       | 25/100    |
|     22           |     37       | 37/100    |
|     23           |     23       | 23/100    |
|    $\sum$        |     100      | 1         |

\\

La distribución marginal segunda quedaría de la siguiente forma:

|     $Y$          |    $n_{.j}$  | $f_{.j}$  |
|------------------|--------------|-----------|
|     75           |     17       |   17/100  |
|    125           |     51       |   51/100  |
|    175           |     32       |   32/100  |
|    $\sum$        |     100      |      1    |


<br>

A las medidas estadísticas (media, varianza,...) calculadas sobre la distribución marginal se les añade el calificativo de marginal (media marginal, varianza marginal, etc.).

### 3.2.2. Distribuciones condicionadas


**Las distribuciones condicionadas expresan como se distribuyen, según una de las dos variables, el conjunto de observaciones que cumplen una condición.** Esta condición viene expresada por un valor o conjunto de valores que presenta la otra variable.

Es decir, la distribución condicionada de $X$ cuando $y$ toma el valor $y_{j}$ o el conjunto de valores $y_{p}$. O bien, cuando la distribución condicionada de $Y$ cuando $x$ toma el valor $x_{i}$ o el conjunto de valores $x_{k}$.


<br>

<font color='Blue'><b> Frecuencia Absoluta Condicionada </b></font>

* Distribución de $X$ condicionada a $Y=y_{j}$, denotada por $X|Y=y_{j}$ es la 
distribución unidimensional de $X$ sabiendo que $Y$ ha tomado la modalidad $y_{j}$ para $\forall j \in N $. Donde $Y$ representa el número de veces que se repite el valor $x_{i}$ teniendo en cuenta solo aquellos valores en que $Y=y_{j}$, así $n_{i|j}=n_{ij}$.

* Distribución de $Y$ condicionada a $X= x_{i}$, denotada por $Y|X=x_{i}$ es la 
distribución unidimensional de $Y$ sabiendo que $X$ ha tomado la modalidad $x_{i}$ para $\forall i \in N $. Donde $X$ representa el número de veces que se repite el valor $y_{j}$ teniendo en cuenta solo aquellos valores en que $X=x_{i}$, así $n_{j|i}=n_{ij}$   

* En  las  distribuciones  condicionadas  no  se  suelen  utilizar  las  distribuciones absolutas, puesto que, estas dependen del número de datos y el número de datos, a su vez, será diferente para cada distribución. Por tanto, dependerá de la frecuencia del valor que se fije  de  la  otra  variable. **Son  mucho  más  útiles las frecuencias  relativas condicionadas**.

<br>

<font color='Blue'><b> Frecuencia Relativa Condicionada </b></font>

* Las distribuciones condicionadas de $X$ dado que $Y = y_{j}$, es decir $X⏐Y=y_{j}$  son:

$ \quad \quad \quad f_{i|j} = \frac{n_{ij}}{n_{.j}} 
\quad \text{donde} \quad 
0 \leq f_{i|j} \leq 1 \quad y \quad \sum_{j=1}^{p}f_{i|j}=1
$

* Las distribuciones condicionadas de $Y$ dado que $X = x_{i}$, es decir $Y⏐X=x_{i}$  son:

$ \quad \quad \quad f_{j|i} = \frac{n_{ij}}{n_{i.}} 
\quad \text{donde} \quad 
0 \leq f_{j|i} \leq 1 \quad y \quad \sum_{i=1}^{k}f_{j|i}=1
$

<br>

$\bbox{Ejemplo.}$ 

Siguiendo el ejemplo visto, una distribución condicionada, sería la distribución según la variable $Y$ (salarios) condicionada a la variable $X$, por ejemplo que la edad sea 21 años.

Es decir, la distribución de la variable $y$ condicionada a que la variable $X$ tome el valor 21, seria: (Y⏐ x= 21). Esto es:

|    $$ Y ⏐ x = 21 $$   | $n_{j}$ $_|$$_{2}$   | $f_{j}$ $_|$$_{2}$ |
|-------------------|----------------------|--------------------|
|    [50  - 100)    |     5                | 5/25    |
|    [100 – 150)    |     15               | 15/25    |
|    [150 – 200)    |     5                | 5/25    |
|    $\sum$         |     25               | 1

\\

Se puede comprobar que, para cada una de las filas de frecuencias de la tabla de correlación se define una distribución condicionada para la variable $y$. Salvo la última que define su distribución marginal. 

Análogamente, cada una de las columnas de frecuencias de la tabla de correlación define una distribución condicionada para la variable $x$, salvo la última que define su distribución marginal.

**Las distribuciones condicionadas son distribuciones unidimensionales** a las cuales se les puede aplicar todo lo conocido para ese tipo de distribuciones.

<br>
<p> <mark>OBSERVACIONES</mark> </p>
<hr>

A las estadísticos calculados sobre las distribuciones condicionadas se les añade el calificativo de *condicionada* (media condicionada, varianza condicionada, etc.).

## 3.4. Independencia y Dependencia estadística


<font color='Blue'><b> Dependencia funcional </b></font>

**La variable $X$ depende, funcionalmente, de la variable $Y$**, si la modalidad que presenta cada individuo en $Y$ determina la que presenta en $X$ (para la modalidad de $Y$ solo hay una modalidad de $X$ con frecuencia no nula). 

Es decir, cuando $X$ depende funcionalmente de $Y$, si a cada valor $y_{j}$ de $Y$ le corresponde un único valor de $X$,  para todo $j \;$ la frecuencia absoluta es$\; n_{ij}=0 $, salvo para un $i$ fijo. 

De esta forma, en cada columna de la tabla de doble entrada sólo hay un término distinto de cero, pero en cada fila pueden figurar varios términos distintos de cero.

<br>

$\bbox{Ejemplo.}$ 

En el siguiente ejemplo puedes observar que, para cada columna, hay solo una frecuencia distinta de cero.  

| X \ Y | 10 | 20 | 30 | 40 |
|---|----|----|----|----|
| 1 | 50 | 0  | 55 | 0  |
| 2 | 0  | 60 | 0  | 0  |
| 3 | 0  | 0  | 0  | 73 |

<br>

Análogamente, **la variable $Y$ depende funcionalmente de la variable $X$** si a  cada modalidad de $X$ le corresponde una sola modalidad de $Y$ con frecuencia no nula. 

Luego, se dice que $Y$ depende funcionalmente de $X$ si a cada valor $x_{i}$ de $X$ le corresponde un único valor de $Y$. Esto ocurre para todo $i$ cuya frecuencia absoluta es $n_{ij}=0$, salvo para un $j$ fijo. 

De esta forma, en cada fila de la tabla de doble entrada sólo hay un término distinto de cero pero en cada columna pueden figurar varios términos distintos de cero. 

<br>

$\bbox{Ejemplo.}$ 

En el siguiente ejemplo se puede observar que en cada fila hay solo una frecuencia distinta de cero. 

| X \ Y | 10 | 20 | 30 | 40 |
|---|----|----|----|----|
| 1 | 0  | 50 | 0  | 0  |
| 2 | 0  | 60 | 0  | 0  |
| 3 | 0  | 0  | 55 | 0  |

<br>

**La dependencia funcional es recíproca** cuando a cada valor $y_{j}$ de $Y$ le corresponde un único valor de $X$ y a cada valor $x_{i}$ de $X$ le corresponde un único valor de $Y$. Así, en cada fila y en cada columna de la tabla de doble entrada sólo hay un número distinto de cero. 

<br>

$\bbox{Ejemplo.}$ 

En el siguiente ejemplo puedes observar que en cada fila hay solo una frecuencia distinta de cero y para cada columna hay solo una frecuencia distinta de cero. 

| X \ Y  | 10 | 20 | 30 | 40 |
|---|----|----|----|----|
| 1 | 40 |    | 0  | 0  |
| 2 | 0  | 60 | 0  | 0  |
| 3 | 0  | 0  | 0  | 55 |
| 4 | 0  | 0  | 73 | 0  |

\\

Consideraciones:

1. Hay que tener en cuenta que la **dependencia funcional no es recíproca**.

2. Los  caracteres  $X$  e  $Y$  son  estadísticamente  dependientes  cuando la variación en uno de ellos influye en la distribución del otro. Luego, la **dependencia estadística es recíproca**.

3. La dependencia funcional no es siempre recíproca, para que sea recíproca es condición necesaria pero no suficiente que la tabla de doble entrada sea cuadrada.

4. Se puede resumir con la siguiente regla: **Dependencia funcional ⇒ Dependencia estadística**.

<font color='Blue'><b> Independencia estadística </b></font>

Se dice que $X$ es **independiente estadísticamente o independiente** del carácter $Y$ si el comportamiento de $X$ es el mismo para todas las modalidades de $Y$. **Esta propiedad es recíproca.** 

Esto es, dos variables aleatorias son independientes si los valores que toma una de ellas no afectan a los de la otra ni a sus probabilidades. Por tanto:

*  El carácter $X$ es independiente estadísticamente o independiente del carácter $Y$ si las  distribuciones de $X$ condicionadas a cualquier modalidad  $y_{j}$ del carácter $Y$ son todas idénticas:

$ \quad \quad f_{i|j} $ no depende de $j$ luego $ f_{i|j} = f_{i}, \; \forall i=1,2,3, \cdots k  $

*  De la misma forma, el carácter $Y$  es independiente estadísticamente o independiente del carácter $X$ si las distribuciones de $Y$ condicionadas a  cualquier modalidad $x_{i}$ del carácter $X$ son todas idénticas. Por tanto:

$ \quad \quad f_{j|i} $ no depende de $i$ luego $ f_{j|i} = f_{j}, \; \forall j=1,2,3, \cdots p  $


Se puede afirmar que si los dos caracteres son independientes entre sí, tanto las filas como las columnas de la tabla de doble entrada serán proporcionales entre sí.

**La condición necesaria y suficiente para que $X$ e $Y$ sean independientes** es que la frecuencia relativa conjunta $f_{ij}$ sea igual al producto de las frecuencias marginales: 

$$ \quad f_{ij} =f_{i.}f_{.j} $$

<p>Al observar dos caracteres en cada individuo, se presenta el problema de determinar la existencia de algún tipo de dependencia entre ellos. En este sentido, hay que destacar los tipos de dependencia que existen:

*	**Dependencia funcional**: Entre dos variables $X$ e $Y$ existe dependencia funcional cuando hay una expresión matemáticas que las relacione. Por ejemplo, los radios de una circunferencia ($X$) y sus longitudes ($Y$).

* **Dependencia aleatoria**: Entre dos variables $X$ e $Y$ existe dependencia aleatoria cuando no existe una expresión matemática que las relacione. Por ejemplo, la edad de los niños (X) y la edad (Y).

* **Existen variables entre las que no existe ningún tipo de dependencia**. Lo que conlleva a decir que los dos conceptos anteriores no son complementarios.

## 3.5. Momentos en distribuciones bidimensionales

Son formulaciones matemáticas, que se definen como **parámetros estadísticos**, algunos de los cuales tienen amplia connotación dentro del estudio de curvas de distribución de frecuencias y, más específicamente, respecto del sesgo y de la curtosis.

Es decir, los **momentos** son unos valores, medidas numéricas, que caracterizan la distribución de frecuencias y **resumen la información recogida en la tabla estadística respecto a alguna propiedad de la variable estadística**.

Los momentos son una **forma de generalizar** toda la teoría relativa a los parámetros estadísticos y guardan relación con una buena parte de ellos. Por ejemplo, algunas medidas de tendencia central y dispersión, como la media y la varianza, son casos particulares de momentos.

Se diferencian normalmente **dos tipos de momentos**: 

* Momentos respecto de la media (centrados).

* Momentos respecto del origen (no centrados).

### 3.5.1. Momentos respecto de la media (centrados)

Se definen los momentos centrales de órdenes $r$ y $s$ como los momentos de órdenes $r$ y $s$ respecto de la media. Se denotan matemáticamente por $\mu_{rs}$.

Su fórmula es:

$$ \bbox[5px,border: 2px solid blue]{ 
   \mu_{rs} = \sum_{i=1}^{k}\sum_{j=1}^{p}(x_{i} - \bar{x})^r (y_{j} - \bar{y})^s f_{ij}
   =\frac{1}{N} \sum_{i=1}^{k}\sum_{j=1}^{p}(x_{i} - \bar{x})^r (y_{j} - \bar{y})^s n_{ij} 
} $$

<br>

Casos particulares:

* Si $r=1$ y $s=0 \Rightarrow \mu_{10} = \sum_{i=1}^{k}\sum_{j=1}^{p}(x_{i} - \bar{x})^1 (y_{j} - \bar{y})^0 f_{ij} = 0 $ 

* Si $r=0$ y $s=1 \Rightarrow \mu_{01} = \sum_{i=1}^{k}\sum_{j=1}^{p}(x_{i} - \bar{x})^0 (y_{j} - \bar{y})^1 f_{ij} = 0 $ 

* Si $r=2$ y $s=0 \Rightarrow \mu_{20} = \sum_{i=1}^{k}\sum_{j=1}^{p}(x_{i} - \bar{x})^2 (y_{j} - \bar{y})^0 f_{ij} = \sigma_{x}^2 $ 

* Si $r=0$ y $s=2 \Rightarrow \mu_{02} = \sum_{i=1}^{k}\sum_{j=1}^{p}(x_{i} - \bar{x})^0 (y_{j} - \bar{y})^2 f_{ij} = \sigma_{y}^2 $ 


Por tanto, se puede resumir en:

* Se cumple la siguiente igualdad: $\mu_{10} = \mu_{01} = 0$.

* Varianzas marginales: $\sigma_{x}^2  = \mu_{20} \;$ y $\; \sigma_{y}^2  = \mu_{02}$.

<br>

**El momento bidimensional respecto a la media más importante** es el momento de órdenes $r=1$ y $s=1$, que recibe el nombre de **Covarianza** y se puede notar por $Cov(x,y)$ o $\sigma_{xy}$. Su caso particular es:

$$\mu_{11} = \sum_{i=1}^{k}\sum_{j=1}^{p}(x_{i} - \bar{x}) (y_{j} - \bar{y}) f_{ij}
=\frac{1}{N} \sum_{i=1}^{k}\sum_{j=1}^{p}(x_{i} - \bar{x}) (y_{j} - \bar{y}) n_{ij} 
$$

<br>

La Covarianza es una medida de la variabilidad conjunta de las variables $X$ e $Y$. **El signo de la covarianza indica el sentido en el que varían conjuntamente** las dos variables. Por tanto:

* Si la $ \mu_{xy} > 0 $ entonces las **variables $X$ e $Y$ varían en el mismo sentido**. Las dos variables crecen o decrecen a la vez (**nube de puntos creciente**). Por ejemplo, cuanto más barato es un bien, más compradores interesados. Otro ejemplo puede ser que a mayor cantidad de agua de lluvia en un año, suele corresponder una mejor cosecha.

* Si la  $ \mu_{xy} < 0 $ entonces las **variables $X$ e $Y$ varían en sentido opuesto**. Cuando una variable crece, la otra variable tiene tendencia a decrecer (**nube de puntos decreciente**). Por ejemplo, a mayor tiempo de espera de un cliente menor satisfacción tendrá el cliente o a mayor renta per cápita en los países suele corresponder una menor mortalidad infantil.

Si **las variables son independientes, la covarianza es nula**. El recíproco no es cierto.

### 3.5.2. Momentos respecto del origen (no centrados)

Se definen los momentos no-centrales de órdenes $r$ y $s$ como los momentos de orden $r$ y $s$ respecto del origen, y se denotan por $m_{rs}$.

Cuya fórmula es:

$$ \bbox[5px,border: 2px solid blue]{ 
   m_{rs} = \sum_{i=1}^{k}\sum_{j=1}^{p}x_{i}^r y_{j}^s f_{ij}
   =\frac{1}{N} \sum_{i=1}^{k}\sum_{j=1}^{p}x_{i}^r y_{j}^s n_{ij} 
} $$

<br>

Casos particulares:

* Si $r=1$ y $s=0 \Rightarrow m_{10} = \sum_{i=1}^{k}\sum_{j=1}^{p}x_{i}^1 y_{j}^0 f_{ij} = \bar{x} $ 

* Si $r=0$ y $s=1 \Rightarrow m_{01} = \sum_{i=1}^{k}\sum_{j=1}^{p}x_{i}^0 y_{j}^1 f_{ij} = \bar{y} $ 


Por tanto, se puede resumir en:

* Medias marginales: $ \bar{x} = m_{10}$ y $\bar{y} = m_{01} $ 

## Actividad: Titanic

Con el dataset de Titanic, se necesita saber lo siguiente:

1. De los pasajeros supervivientes, ¿cuántos pertenecen a la primera clase?

2. ¿De qué clase socio-económica fueron los que más murieron a bordo del Titanic?

3. De los pasajeros de la primera clase, ¿se observa alguna relación entre el precio del billete y la edad?

4. De los pasajeros que no sobrevivieron de primera clase, ¿se observa alguna relación entre el precio del billete y la edad? ¿y de los que si sobrevivieron?

### Solución 

[*A continuación, haz clic para conocer la solución.*]

Primero debemos cargar el dataset para, después, comenzar con la resolución de la actividad.

```
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Carga del fichero desde el enlace web y creación del dataframe
url_data = 'https://raw.githubusercontent.com/md-lorente/data/master/titanic.csv'

# Creación Dataframe
df = pd.read_csv(url_data, sep=',')

# Visualización del dataframe (la cabecera)
df.head()
```

```
# Tabla de doble entrada
tabla = pd.crosstab(df["Pclass"],df["Survived"] , margins=True) 
tabla
```

```
tot_clase_1 = tabla[1][:-1].idxmax()
num_clase_1 = tabla[1][:-1].max()
print("Los pasajeros que más sobrevivieron fueron los de " +str(tot_clase_1), "ª clase, con un total de", num_clase_1, "pasajeros" )
```

```
tot_clase_0 = tabla[0][:-1].idxmax()
num_clase_0 = tabla[0][:-1].max()
print("Los pasajeros que más murieron fueron los de " +str(tot_clase_0), "ª clase con un total de", num_clase_0, "pasajeros" )
```

```
# # Copy df profunda y filtrando por clase socio-económica
## filtramos las observaciones aplicando un filtro por fila
df_1class = df[df['Pclass'] == 1].copy(deep='True')

df_1class = pd.DataFrame(df_1class)

# Diagrama de dispersión o nube de puntos
## etiqueta
plt.title("Dispersión del precio del billete y la edad de los pasajeros de 1ª clase")
plt.xlabel("Edad de los pasajeros")
plt.ylabel("Precio del billete")

plt.scatter(df_1class["Age"],df_1class["Fare"])
plt.show()
```

No se observa ninguna relación entre la edad de los pasajeros y el precio del billete.

```
# Copy df profunda y filtrando
## filtramos las observaciones aplicando un filtro por fila
df_1class_s0 = df[(df['Survived'] == 0) & (df['Pclass'] == 1) ].copy(deep='True')
df_1class_s0 = pd.DataFrame(df_1class_s0)

## Nube de puntos
plt.title("Dispersión del precio del billete y la edad de los pasajeros de 1ª clase que no sobrevivieron")
plt.xlabel("Edad de los pasajeros")
plt.ylabel("Precio del billete")

plt.scatter(df_1class_s0["Age"],df_1class_s0["Fare"])
plt.show()
```

```
# Copy df profunda y filtrando
## NO sobreviven de la primera clase
df_1class_s0 = df[(df['Survived'] == 0) & (df['Pclass'] == 1) ].copy(deep='True')
df_1class_s0 = pd.DataFrame(df_1class_s0)

## SI sobreviven de la primera clase
df_1class_s1 = df[(df['Survived'] == 1) & (df['Pclass'] == 1) ].copy(deep='True')
df_1class_s1 = pd.DataFrame(df_1class_s1)


## Nube de puntos
plt.suptitle("Precio del billete VS edad pasajeros de 1ª clase", fontsize=14)
plt.ylim(0, 500)
plt.xlim(0, 90)

### Nube de puntos 1
plt.subplot(121)
plt.title("SOBREVIVIERON")
plt.xlabel("Edad de los pasajeros")
plt.ylabel("Precio del billete")
plot1 = plt.scatter(df_1class_s1["Age"],df_1class_s1["Fare"])

### Nube de puntos 2
plt.subplot(122)
plt.title("NO SOBREVIVIERON")
plt.xlabel("Edad de los pasajeros")
plot2 = plt.scatter(df_1class_s0["Age"],df_1class_s0["Fare"])

### Comparativa
plt.show(plot1, plot2)
```

De manera intuitiva, se puede apreciar claramente que las variables **no muestran ningún tipo de relación** entre la variable precio del billete y la variable edad de los pasajero. Cada observación se muestra sin "ninguna forma determinada".
